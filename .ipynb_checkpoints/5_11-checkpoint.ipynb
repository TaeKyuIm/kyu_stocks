```python
from pykrx import stock
from kyu_stocks import *
market = "KOSDAQ"
tickers = stock.get_market_ticker_list(market=market)
```

```python
import time
```

```python
from tqdm import tqdm

week = []
for ticker in tqdm(tickers):
    df = get_52price(ticker, "20220501", "20230511")
    if df['종가'].iloc[-1] > 0.9 * df.iloc[-1, -2]:
        week.append([ticker, df.iloc[-1, -2]])
    time.sleep(2)
```

```python
print(week)
```

```python
import pickle
with open('./90_candidate_2.pickle', 'wb') as f:
    pickle.dump(week, f)
```
